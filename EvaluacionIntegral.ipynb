## Descripcion del Dataset

#### El dataset Diamonds.csv contiene información detallada sobre diferentes diamantes. A continuación se describe cada columna del dataset


- __carat__: Peso del diamante en quilates. Es una variable numérica continua.
- __cut__: Calidad del corte del diamante. Es una variable categórica que puede tomar los siguientes valores:
    - Fair
    - Good
    - Very Good
    - Premium
    - Ideal
- __color__: Color del diamante, clasificado de acuerdo a la escala de letras del D al J, donde D representa el mejor color (incoloro) y J el peor color (ligeramente teñido). Es una variable categórica.
- __clarity__: Claridad del diamante, medida de acuerdo a la cantidad y visibilidad de inclusiones y defectos. Es una variable categórica con los siguientes posibles valores:
    - FL (Flawless)
    - IF (Internally Flawless)
    - VVS1 (Very Very Slightly Included 1)
    - VVS2 (Very Very Slightly Included 2)
    - VS1 (Very Slightly Included 1)
    - VS2 (Very Slightly Included 2)
    - SI1 (Slightly Included 1)
    - SI2 (Slightly Included 2)
- __depth__: Profundidad total del diamante como un porcentaje del promedio del diámetro. Es una variable numérica continua.
- __table__: Tamaño de la mesa del diamante (parte superior plana) como un porcentaje del diámetro del promedio. Es una variable numérica continua.
- __price__: Precio del diamante en dólares estadounidenses. Es una variable numérica continua.
- __x__: Longitud del diamante en milímetros. Es una variable numérica continua.
- __y__: Ancho del diamante en milímetros. Es una variable numérica continua.
- __z__: Altura del diamante en milímetros. Es una variable numérica continua.

## Importacion de librerias necesarias para el analisis

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

## Carga del dataset

```python
df = pd.read_csv('diamonds.csv')
```

## Crear una nueva columna para clasificar los precios en rangos

```python
# Crear categorías de precios
bins = [0, 500, 1000, 5000, 10000, 20000]
labels = ['Bajo', 'Medio-Bajo', 'Medio', 'Medio-Alto', 'Alto']
df['price_category'] = pd.cut(df['price'], bins=bins, labels=labels)
```

## Evaluacion

```python
# Creamos dos grafico de dispersion, uno con hue y otro sin hue.
# Lo hacemos con seaborn para que tengas la comparacion entre ambos
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Precio de los diamantes sin hue')
plt.show()

sns.scatterplot(x='carat', y='price', hue='clarity', data=df)
plt.title('Precio de los diamantes con hue')
plt.show()
```

```python
# Creo un histograma para ver la distribucion de los precios de los diamantes
sns.histplot(df['price'], bins=100, kde=True)
plt.title('Distribución de precios de los diamantes')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Creo un Scatter para ver cual es el correcto. Entre x y carats
sns.scatterplot(x='x', y='price', data=df)
plt.title('Relación entre Carat y Precio')
plt.xlabel('Carat')
plt.ylabel('Precio')
plt.show()
```

```python
# Creo un boxplot para ver la profundidad del diamante por claridad
sns.boxplot(x='clarity', y='depth', data=df)
plt.title('Profundidad del diamante por claridad')
plt.xlabel('Claridad')
plt.ylabel('Profundidad')
plt.show()
```

```python
# Creo un grafico de barras para mostrar la distribucion de la categoria de precios
sns.countplot(x='price_category', data=df, order=labels)
plt.title('Distribución de la categoría de precios')
plt.xlabel('Categoría de precios')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Creo el grafico de dispersion para ver la relacion entre tabla y depth
sns.scatterplot(x='table', y='depth', data=df, alpha=0.5)
plt.title('Relación entre Tabla y Profundidad')
plt.xlabel('Tabla')
plt.ylabel('Profundidad')
plt.show()
```

```python
# creo un grafico de violin para ver la distribucion de los precios por categoria de corte
sns.violinplot(x='cut', y='price', hue='cut', data=df, palette='viridis')
plt.title('Distribución de precios por categoría de corte')
plt.xlabel('Categoría de corte')
plt.ylabel('Precio')
plt.show()
```

```python
# Creo un scatterplot para ver relacion entre carat, depth y color
sns.scatterplot(x='carat', y='depth', hue='color', data=df, alpha=0.7)
plt.title('Relación entre Carat, Depth y Color')
plt.xlabel('Carat')
plt.ylabel('Depth')
plt.show()
```

```python
# Creo el grafico de barras apiladas que muestra la distribucionde la categoria de precios por cut
price_cut_counts = df.groupby(['cut', 'price_category']).size().unstack(fill_value=0)
price_cut_counts.plot(kind='bar', stacked=True)
plt.title('Distribución de la categoría de precios por corte')
plt.xlabel('Categoría de corte')
plt.ylabel('Frecuencia')
plt.legend(title='Categoría de precios')
plt.show()
```
