### Personalización de la estética con Seaborn
1. Temas en Seaborn


Seaborn proporciona varios temas predefinidos que podemos usar para cambiar la apariencia de nuestros gráficos. Los temas disponibles son: darkgrid, whitegrid, dark, white, y ticks.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Establecer el tema
sns.set_theme(style="darkgrid")

# Crear un gráfico de dispersión
data = sns.load_dataset("tips")
sns.scatterplot(data=data, x="total_bill", y="tip", hue="day")
plt.show()
```

2. Paletas de colores
* Podemos personalizar los colores de nuestros gráficos utilizando las paletas predefinidas de Seaborn o creando nuestras propias paletas. Algunas paletas comunes son: deep, muted, pastel, bright, dark, y colorblind. 





```python
# List of Seaborn palette names
palettes = ["deep", "muted", "bright", "pastel", "dark", "colorblind",
            "Blues", "BuGn", "BuPu", "GnBu", "Greens", "Greys", "Oranges", "OrRd", "PuBu", "PuBuGn", "PuRd", "Purples", "RdPu", "Reds", "YlGn", "YlGnBu", "YlOrBr", "YlOrRd",
            "afmhot", "autumn", "binary", "bone", "cool", "copper", "cubehelix", "flag", "gist_earth", "gist_gray", "gist_heat", "gist_ncar", "gist_rainbow", "gist_stern", "gist_yarg", "gnuplot", "gnuplot2", "gray", "hot", "hsv", "inferno", "magma", "nipy_spectral", "ocean", "pink", "plasma", "prism", "rainbow", "seismic", "spring", "summer", "tab10", "tab20", "tab20b", "tab20c", "terrain", "viridis", "winter"]
```

```python
sns.color_palette('gist_earth')
```

```python

# Plot all the palettes
n = len(palettes)
fig, axes = plt.subplots(ncols=1, nrows=n, figsize=(8, 2*n), dpi=80)

for ax, palette in zip(axes, palettes):
    colors = sns.color_palette(palette)
    n_colors = len(colors)
    ax.imshow([colors], aspect='auto')
    ax.set_xticks(range(n_colors))
    ax.set_xticklabels(range(1, n_colors+1))
    ax.set_yticks([])
    ax.set_title(palette, fontsize=12, pad=5)

plt.tight_layout()
plt.show()
```

```python
sns.set_theme(style="whitegrid", palette="terrain")

data = sns.load_dataset("iris")
sns.boxplot(data=data, x="species", y="sepal_length")
plt.show()
```

3. Tema oscuro y paleta "pastel":

```python
sns.set_theme(style="dark", palette="viridis")

data = sns.load_dataset("penguins")
sns.violinplot(data=data, x="species", y="bill_length_mm", hue="sex", split=True)
plt.show()
```

4. Seaborn permite ajustar el estilo general de los gráficos, incluyendo elementos como el tipo de cuadrícula y el fondo.

* Estilo sin cuadrícula y paleta "bright":



```python
sns.set_theme(style="ticks", palette="bright")

data = sns.load_dataset("flights")
sns.lineplot(data=data, x="year", y="passengers", hue="month")
plt.show()
```

El contexto ajusta la escala de los elementos gráficos para diferentes contextos de presentación: paper, notebook, talk, y poster.

```python
sns.set_context("paper")

# Crear un gráfico de línea
flights = sns.load_dataset("flights")
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.show()
```

5. Paleta de color personalizada:



```python
custom_palette = sns.color_palette(["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"])
sns.set_theme(palette=custom_palette)

data = sns.load_dataset("diamonds")
sns.scatterplot(data=data, x="carat", y="price", hue="cut")
plt.show()
```

6. Estilo de cuadrícula blanca y fondo con imagen:



```python
import numpy as np
from matplotlib.colors import ListedColormap

sns.set_theme(style="whitegrid")
data = sns.load_dataset("titanic")

cmap = ListedColormap(sns.color_palette("coolwarm", 256).as_hex())
sns.kdeplot(data=data, x="age", y="fare", cmap=cmap, fill=True)
plt.show()
```

7. Paleta "cubehelix" y estilo sin fondo:



```python
sns.set_theme(style="ticks", palette="cubehelix")

data = sns.load_dataset("tips")
sns.histplot(data=data, x="total_bill", hue="sex", element="step")
plt.show()
```

8. Tema blanco y paleta "Set2":



```python
sns.set_theme(style="white", palette="Set2")

data = sns.load_dataset("mpg")
sns.regplot(data=data, x="horsepower", y="mpg", scatter_kws={"s": 50}, line_kws={"color": "red"})
plt.show()
```

9. Estilo "dark" con ajuste de contexto:

```python
sns.set_theme(style="dark", palette="colorblind", context="talk")

data = sns.load_dataset("planets")
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x="year", hue="method", multiple="stack")
plt.show()
```

10. Paleta "viridis" y estilo con cuadrícula menor:

```python
sns.set_theme(style="whitegrid", palette="viridis")

data = sns.load_dataset("exercise")
sns.barplot(data=data, x="diet", y="pulse", hue="time")
plt.show()
```
