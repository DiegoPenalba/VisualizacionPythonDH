### Visualización univariante
#### kdeplot: Trazados de Densidad más Flexibles y Configurables

Importo las librerías necesarias

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
```

Traemos el dataset 'tips' de Seaborn

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

Ejemplo 1: Creamos un gráfico KDE básico para la variable total_bill, con sombreado (shade=True) y color skyblue.

```python


# Ejemplo 1: KDE básico
plt.figure(figsize=(10, 6))
sns.kdeplot(tips['total_bill'], shade=True, color='skyblue')
plt.title('Estimación de Densidad del Núcleo (KDE) del Total de la Cuenta')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Densidad')
plt.show()

```

Ejemplo 2: Ajustamos el ancho de banda con bw_adjust=0.5 para obtener una estimación más detallada.



```python

# Ejemplo 2: Ajuste del Ancho de Banda (bw_adjust)
plt.figure(figsize=(10, 6))
sns.kdeplot(tips['total_bill'], shade=True, color='skyblue', bw_adjust=0.5)
plt.title('KDE del Total de la Cuenta con Ancho de Banda Ajustado')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Densidad')
plt.show()

```

Ejmeplo 3: Creamos un gráfico KDE acumulativo utilizando cumulative=True.

```python
# Ejemplo 3: KDE Acumulativo
plt.figure(figsize=(10, 6))
sns.kdeplot(tips['total_bill'], shade=True, color='skyblue', cumulative=True)
plt.title('Estimación de Densidad Acumulativa del Total de la Cuenta')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Densidad Acumulada')
plt.show()
```

Ejemplo 4: Creamos un gráfico KDE bivariado para visualizar la relación entre total_bill y tip con sombreado y un mapa de color Blues.

```python

# Ejemplo 4: KDE Bivariado
plt.figure(figsize=(10, 6))
sns.kdeplot(x=tips['total_bill'], y=tips['tip'], cmap='Blues', shade=True)
plt.title('Estimación de Densidad Bivariada del Total de la Cuenta y Propinas')
plt.xlabel('Total de la Cuenta ($)')
plt.ylabel('Propina ($)')
plt.show()
```
