### Jointplots
Importamos librerías y cargamos el dataset Tips


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset 'tips' de Seaborn
tips = sns.load_dataset('tips')

```

Creamos el jointplot

```python
# Crear un jointplot
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', marginal_kws=dict(bins=15, fill=True))

# Mostrar el gráfico
plt.show()
```

#### Tipos de gráficos con jointplot

1. Scatter Plot (Diagrama de Dispersión)

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

2. Regression Plot (Diagrama de Regresión)

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

3.KDE Plot (Gráfico de KDE)

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

4. Hexbin Plot (Gráfico de Hexágonos)

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

Personalización

Se personalizan los histogramas marginales y se utiliza un gráfico de regresión en el centro

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg', marginal_kws=dict(bins=20, kde=True))

plt.show()
```
