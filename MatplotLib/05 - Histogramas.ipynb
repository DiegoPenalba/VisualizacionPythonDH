### Creación de Histogramas Básicos

Importación de Bibliotecas Necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de Datos de Ejemplo

```python
np.random.seed(0)
datos = np.random.randn(1000)
```

Creación de un Histograma Básico


```python
plt.hist(datos, bins=30, color='blue', edgecolor='black')
plt.title('Histograma Básico')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

### Representación de Distribuciones y Frecuencias
Ajuste del Número de Bins

```python
plt.hist(datos, bins=15, color='green', edgecolor='black')
plt.title('Histograma con 50 Bins')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Normalización de Histogramas

```python
plt.hist(datos, bins=30, density=True, color='purple', edgecolor='black')
plt.title('Histograma Normalizado')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

### Personalización de Histogramas
Cambiar Colores y Estilos

```python
plt.hist(datos, bins=30, color='orange', edgecolor='black', linestyle='--', linewidth=1.2)
plt.title('Histograma Personalizado')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Añadir Títulos y Etiquetas


```python
plt.hist(datos, bins=30, color='red', edgecolor='black')
plt.title('Histograma con Títulos y Etiquetas')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Personalización de Ejes y Leyendas


```python
plt.hist(datos, bins=30, color='cyan', edgecolor='black', label='Datos')
plt.title('Histograma con Leyenda')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

### Técnicas Avanzadas en Histogramas
Superposición de Múltiples Histogramas

```python
datos1 = np.random.randn(1000)
datos2 = np.random.randn(1000) + 2

plt.hist(datos1, bins=30, alpha=0.5, label='Datos 1', color='blue', edgecolor='black')
plt.hist(datos2, bins=30, alpha=0.5, label='Datos 2', color='red', edgecolor='black')

plt.title('Superposición de Múltiples Histogramas')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

Histogramas con Densidad de Kernel (KDE)

```python
import seaborn as sns

sns.histplot(datos, bins=30, kde=True, color='green')
plt.title('Histograma con Densidad de Kernel (KDE)')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Subplots con Histogramas


```python
datos1 = np.random.randn(1000)
datos2 = np.random.randn(1000) + 2

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].hist(datos1, bins=30, color='blue', edgecolor='black')
axs[0].set_title('Histograma 1')
axs[0].set_xlabel('Valores')
axs[0].set_ylabel('Frecuencia')

axs[1].hist(datos2, bins=30, color='red', edgecolor='black')
axs[1].set_title('Histograma 2')
axs[1].set_xlabel('Valores')
axs[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

### Ejemplos Prácticos
Ejemplo Básico

```python
plt.hist(datos, bins=30, color='blue', edgecolor='black')
plt.title('Ejemplo Básico de Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

Ejemplo Avanzado con Personalización


```python
plt.hist(datos, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.axvline(datos.mean(), color='red', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(datos.mean()*1.1, max_ylim*0.9, 'Media: {:.2f}'.format(datos.mean()), color='red')
plt.title('Ejemplo Avanzado de Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
