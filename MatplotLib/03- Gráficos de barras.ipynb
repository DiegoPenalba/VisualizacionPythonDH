### Importación de librerias necesarias

```python
import matplotlib.pyplot as plt
import numpy as np
```

Generación de Datos de Ejemplo

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [5, 7, 3, 8, 6]
```

Creación de un Gráfico de Barras Verticales

```python
plt.bar(categorias, valores, color='skyblue')
plt.title('Gráfico de Barras Verticales')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

Modificación del Gráfico de Barras Verticales a Horizontales

```python
plt.barh(categorias, valores, color='salmon')
plt.title('Gráfico de Barras Horizontales')
plt.xlabel('Valores')
plt.ylabel('Categorías')
plt.show()
```

Generación de Datos para Barras Apiladas

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [5, 7, 3, 8, 6]
valores2 = [2, 4, 6, 1, 3]
```

Creación de un Gráfico de Barras Apiladas

```python
bar_width = 0.35

# Creación de las barras
plt.bar(categorias, valores1, color='skyblue', label='Valor 1')
plt.bar(categorias, valores2, bottom=valores1, color='salmon', label='Valor 2')

plt.title('Gráfico de Barras Apiladas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.show()
```

Personalización de Gráficos de Barras

```python
plt.bar(categorias, valores, color='purple', edgecolor='black', hatch='//')
plt.title('Gráfico de Barras Personalizado')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

Personalización de Ejes y Leyendas

```python
plt.bar(categorias, valores1, color='blue', label='Grupo 1')
plt.bar(categorias, valores2, bottom=valores1, color='red', label='Grupo 2')

plt.title('Gráfico de Barras Apiladas con Leyenda')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.grid(True)
plt.show()
```

Ejemplo Básico

```python
plt.bar(categorias, valores, color='skyblue')
plt.title('Ejemplo Básico de Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

Ejemplo Avanzado con Personalización

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [5, 7, 3, 8, 6]
valores2 = [2, 4, 6, 1, 3]
bar_width = 0.35

plt.bar(categorias, valores1, color='blue', edgecolor='black', label='Grupo 1')
plt.bar(categorias, valores2, bottom=valores1, color='orange', edgecolor='black', label='Grupo 2', hatch='//')

plt.title('Ejemplo Avanzado de Gráfico de Barras Apiladas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
