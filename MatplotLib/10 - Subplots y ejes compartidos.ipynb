### Creación de Subplots Básicos
Uso de plt.subplot para Crear Subplots

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 1, 2, 3, 4]

plt.subplot(2, 1, 1)  # (nrows, ncols, index)
plt.plot(x, y1)
plt.title('Gráfico 1')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Gráfico 2')

plt.tight_layout()
plt.show()
```

Uso de plt.subplots para Crear una Cuadrícula de Subplots


```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Gráfico 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Gráfico 2')

axs[1, 0].plot(x, y1)
axs[1, 0].set_title('Gráfico 3')

axs[1, 1].plot(x, y2)
axs[1, 1].set_title('Gráfico 4')

plt.tight_layout()
plt.show()
```

### Personalización de Subplots
Ajuste de Espaciado entre Subplots

```python
fig, axs = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.4, wspace=0.4)  # Ajuste de espacio horizontal y vertical

for i in range(2):
    for j in range(2):
        axs[i, j].plot(x, y1)
        axs[i, j].set_title(f'Gráfico {i * 2 + j + 1}')

plt.show()
```

Añadir Títulos y Etiquetas a Subplots Individuales


```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Gráfico 1')
axs[0, 0].set_xlabel('Eje X')
axs[0, 0].set_ylabel('Eje Y')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Gráfico 2')
axs[0, 1].set_xlabel('Eje X')
axs[0, 1].set_ylabel('Eje Y')

axs[1, 0].plot(x, y1)
axs[1, 0].set_title('Gráfico 3')
axs[1, 0].set_xlabel('Eje X')
axs[1, 0].set_ylabel('Eje Y')

axs[1, 1].plot(x, y2)
axs[1, 1].set_title('Gráfico 4')
axs[1, 1].set_xlabel('Eje X')
axs[1, 1].set_ylabel('Eje Y')

plt.tight_layout()
plt.show()
```

### Ejes Compartidos
Creación de Subplots con Ejes Compartidos

```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Gráfico 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Gráfico 2')

axs[1, 0].plot(x, y1)
axs[1, 0].set_title('Gráfico 3')

axs[1, 1].plot(x, y2)
axs[1, 1].set_title('Gráfico 4')

plt.tight_layout()
plt.show()
```

Sincronización de los Ejes en Subplots


```python
fig, axs = plt.subplots(2, 2, sharex='col', sharey='row')

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Gráfico 1')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Gráfico 2')

axs[1, 0].plot(x, y1)
axs[1, 0].set_title('Gráfico 3')

axs[1, 1].plot(x, y2)
axs[1, 1].set_title('Gráfico 4')

plt.tight_layout()
plt.show()
```

### Ejemplos Prácticos
Ejemplo Básico de Subplots

```python
fig, axs = plt.subplots(1, 2)

axs[0].plot(x, y1)
axs[0].set_title('Gráfico 1')

axs[1].plot(x, y2)
axs[1].set_title('Gráfico 2')

plt.tight_layout()
plt.show()
```

Ejemplo Avanzado con Subplots y Ejes Compartidos


```python
fig, axs = plt.subplots(2, 2, sharex=True, sharey=True)
fig.suptitle('Ejemplo Avanzado con Subplots y Ejes Compartidos')

for i in range(2):
    for j in range(2):
        axs[i, j].plot(x, y1 if (i + j) % 2 == 0 else y2)
        axs[i, j].set_title(f'Gráfico {i * 2 + j + 1}')

plt.tight_layout()
plt.show()
```
